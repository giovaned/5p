```python
# draw_rectangle.py

from PIL import Image, ImageDraw


def rectangle(output_path):
    image = Image.new("RGBA", (1200, 1200), (102,204,255,255))
    draw = ImageDraw.Draw(image)
    draw.rectangle((90, 180, 830, 185), fill=(153,153,153,255))
    out = Image.alpha_composite(image, image)
    image.save(output_path)
    image.show()

rectangle("rectangle.png")
```

```python
from PIL import ImageFont
file = "rectangle.png"
img = Image.open(file)
title_font = ImageFont.truetype('fonts/EncodeReg.ttf', 67)
author_text = "Author & Second Author"
title_text = "The great title here"
image_editable = ImageDraw.Draw(img)
image_editable.text((100,80), author_text, (0, 0, 0), font=title_font)
image_editable.text((100,200), title_text, (0, 0, 0), font=title_font)
img.save('orange-cat.png')
```
